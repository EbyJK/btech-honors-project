```python
import shutil
import kagglehub
dataset_path = kagglehub.dataset_download(
    "rupeshkumaryadav/mumbai-slum-detection-dataset"
)
CONTENT_EXTRACT_PATH = "/content/mumbai-slum-detection-dataset"
# Make working copy in /content (this is what you train on)
shutil.copytree(dataset_path, CONTENT_EXTRACT_PATH, dirs_exist_ok=True)
print("Working dataset available at:", CONTENT_EXTRACT_PATH)
```

```python
# =========================================================
# Patch-level Water Classification from Full Raster Data
# (Sentinel bands + pixel labels)
# =========================================================

import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn.model_selection import train_test_split

# ---------- Paths ----------
DATASET_ROOT = "/content/mumbai-slum-detection-dataset/Data"
X_PATH = os.path.join(DATASET_ROOT, "sentinel_input.npy")
Y_PATH = os.path.join(DATASET_ROOT, "slum_labels.npy")

# ---------- Load Data ----------
print("📥 Loading raster data...")
X = np.load(X_PATH)     # (bands, H, W)
y = np.load(Y_PATH)     # (H, W)

print("X:", X.shape)
print("y:", y.shape)

# ---------- Rearrange Channels ----------
X = np.transpose(X, (1, 2, 0))  # (H, W, C)

# ---------- Normalize ----------
X = X.astype("float32")
X = (X - X.min()) / (X.max() - X.min() + 1e-8)

# ---------- Patch Extraction ----------
PATCH_SIZE = 32
STRIDE = 32

patches = []
labels = []

for i in range(0, X.shape[0] - PATCH_SIZE, STRIDE):
    for j in range(0, X.shape[1] - PATCH_SIZE, STRIDE):
        patch = X[i:i+PATCH_SIZE, j:j+PATCH_SIZE, :]
        mask_patch = y[i:i+PATCH_SIZE, j:j+PATCH_SIZE]

        # Label rule: slum if >30% pixels are slum
        label = 1 if np.mean(mask_patch) > 0.3 else 0

        patches.append(patch)
        labels.append(label)

patches = np.array(patches)
labels = np.array(labels)

print("Total patches:", patches.shape[0])

# ---------- Train / Validation Split ----------
X_train, X_val, y_train, y_val = train_test_split(
    patches, labels,
    test_size=0.2,
    random_state=42,
    stratify=labels
)

print("Train patches:", X_train.shape)
print("Val patches:", X_val.shape)

# ---------- CNN Model ----------
model = models.Sequential([
    layers.Conv2D(32, 3, activation="relu", input_shape=patches.shape[1:]),
    layers.MaxPooling2D(),

    layers.Conv2D(64, 3, activation="relu"),
    layers.MaxPooling2D(),

    layers.Conv2D(128, 3, activation="relu"),
    layers.MaxPooling2D(),

    layers.Flatten(),
    layers.Dense(128, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# ---------- Callbacks ----------
checkpoint = ModelCheckpoint(
    "/content/best_slum_patch_classifier.h5",
    monitor="val_accuracy",
    save_best_only=True,
    mode="max"
)

early_stop = EarlyStopping(
    patience=5,
    restore_best_weights=True
)

# ---------- Train ----------
print("🚀 Training started...")
model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=20,
    batch_size=32,
    callbacks=[checkpoint, early_stop]
)

# ---------- Save Final Model ----------
model.save("/content/slum_patch_classifier_final.h5")
print("✅ Model saved")

# ---------- Visual Example ----------
idx = 0
pred = model.predict(X_val[idx:idx+1])[0][0]
label = "Water" if pred > 0.5 else "Non-Water"

plt.figure(figsize=(4,4))
plt.imshow(X_val[idx][:,:,0], cmap="gray")
plt.title(f"Prediction: {label} ({pred:.2f})")
plt.axis("off")
plt.show()
```

```python
# =========================================================
# Model Evaluation: How good is the Water classifier?
# =========================================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    roc_auc_score,
    roc_curve
)

# ---------- Predict on validation set ----------
y_prob = model.predict(X_val).ravel()
y_pred = (y_prob > 0.5).astype(int)

# ---------- Class balance ----------
print("Class distribution in validation set:")
print("Non-Water (0):", np.sum(y_val == 0))
print("Water (1):    ", np.sum(y_val == 1))

# ---------- Confusion Matrix ----------
cm = confusion_matrix(y_val, y_pred)
tn, fp, fn, tp = cm.ravel()

print("\nConfusion Matrix:")
print(cm)

# ---------- Metrics ----------
precision = tp / (tp + fp + 1e-8)
recall    = tp / (tp + fn + 1e-8)
f1        = 2 * precision * recall / (precision + recall + 1e-8)

print(f"\nPrecision: {precision:.3f}")
print(f"Recall:    {recall:.3f}")
print(f"F1-score:  {f1:.3f}")

# ---------- Full Classification Report ----------
print("\nClassification Report:")
print(classification_report(y_val, y_pred, target_names=["Non-Water", "Water"]))

# ---------- ROC-AUC ----------
auc = roc_auc_score(y_val, y_prob)
print(f"ROC-AUC: {auc:.3f}")

# ---------- ROC Curve ----------
fpr, tpr, _ = roc_curve(y_val, y_prob)

plt.figure(figsize=(5,5))
plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve – Water Classifier")
plt.legend()
plt.show()
```

```python
# =========================================================
# Run trained water patch classifier and overlay prediction on input image
# =========================================================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from pathlib import Path

# ---------- Paths ----------
IMG_PATH = "/content/map-screenshot(2).png"
MODEL_PATH = "/content/drive/MyDrive/Honours/slum_patch_classifier_final.h5"
PATCH_SIZE = 32
THRESHOLD = 0.01  # hide low-confidence areas from overlay

# ---------- Load image ----------
img = cv2.imread(IMG_PATH, cv2.IMREAD_GRAYSCALE)
if img is None:
    raise FileNotFoundError(f"Image not found at {IMG_PATH}")

# Crop to patch-aligned size
h, w = img.shape
h = (h // PATCH_SIZE) * PATCH_SIZE
w = (w // PATCH_SIZE) * PATCH_SIZE
img = img[:h, :w]

# Normalize image
img = img.astype("float32")
img = (img - img.min()) / (img.max() - img.min() + 1e-8)

# Convert to 4-channel input (change to 3-channel if needed)
img_4ch = np.stack([img]*4, axis=-1)  # shape: (H, W, 4)

# ---------- Extract patches ----------
patches = []
coords = []
for i in range(0, h, PATCH_SIZE):
    for j in range(0, w, PATCH_SIZE):
        patch = img_4ch[i:i+PATCH_SIZE, j:j+PATCH_SIZE, :]
        patches.append(patch)
        coords.append((i, j))
patches = np.array(patches)

# ---------- Load model ----------
model = load_model(MODEL_PATH)

# ---------- Predict ----------
probs = model.predict(patches, batch_size=64).ravel()

# ---------- Reconstruct probability map ----------
prob_map = np.zeros((h, w))
for (i, j), p in zip(coords, probs):
    prob_map[i:i+PATCH_SIZE, j:j+PATCH_SIZE] = p

# ---------- Apply threshold (optional) ----------
overlay = prob_map.copy()
overlay[overlay < THRESHOLD] = np.nan  # mask low-probability

# ---------- Enhanced Overlay Visualization ----------
plt.figure(figsize=(10, 10))
plt.imshow(img, cmap="gray")
plt.imshow(overlay, cmap="hot", alpha=0.7)
plt.colorbar(label="Water Probability")
plt.title("Enhanced Water Prediction Overlay")
plt.axis("off")
plt.show()
```
